# **Assignment on Uber dataset**

**Load the dataset**

```python
import pandas as pd
uber_data=pd.read_csv("Uber.csv")
print(uber_data)
```

**Display basic info about dataset**

```python
uber_data.info()
```

**Check for missing values**

```python
missing_values = uber_data.isnull().sum()
print(missing_values)
```

**Drop rows with missing values**

```python
Drop_rows = uber_data.dropna()
print("\nData after dropping rows with missing values:")
print(Drop_rows)
```

**fill missing values (propose column with unknown value)**

```python
uber_data.loc[:, 'PURPOSE*'] = uber_data['PURPOSE*'].fillna('Unknown')
```

**Check and remove duplicates**

```python
duplicates=uber_data.duplicated()
duplicates
```

```python
uber_data=uber_data.drop_duplicates()
uber_data
```

**Convert START_DATE and END_DATE to datetime**

```python
uber_data['START_DATE*']=pd.to_datetime(uber_data['START_DATE*'],errors='coerce')
```

```python
uber_data['END_DATE*']=pd.to_datetime(uber_data['END_DATE*'],errors='coerce')
uber_data.info()
```

**Total number of rides per category:**

```python
ride_counts = uber_data.groupby('CATEGORY*').size()
print(ride_counts)
```

**Total miles traveled for each purpose:**

```python
total_miles= uber_data.groupby('PURPOSE*')['MILES*'].sum()
print(total_miles)
```

**Average distance for business vs. personal rides:**

```python
average_distance = uber_data.groupby('CATEGORY*')['MILES*'].mean()
print(average_distance)
```

**Add a column for cost estimation (assuming $2 per mile):**

```python
uber_data['COST_ESTIMATION'] = uber_data['MILES*'] * 2
uber_data
```

**Filter rides longer than 50 miles:**

```python
filtered_rides=uber_data[uber_data['MILES*']>50]
filtered_rides
```

**Filter by specific purpose (e.g., meetings):**

```python
filtered_rides = uber_data[uber_data['PURPOSE*'] == 'Meeting']
print(filtered_rides)
```

**What is the total number of business trips versus personal trips?**

```python
trip = uber_data.groupby('CATEGORY*').size()
trip
```

**What percentage of trips are business versus personal?**

```python
trip=uber_data['CATEGORY*'].value_counts()
```

```python
trip_percentage = (trip / len(uber_data)) * 100
print(trip_percentage)
```
