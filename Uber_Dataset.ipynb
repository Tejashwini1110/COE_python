```python
uber_data=pd.read_csv("Uber.csv")
uber_data.head()
```

```python
filtered_data=uber_data[uber_data['MILES*']>50]
print(filtered_data)
```

```python
filtered_data=uber_data[(uber_data['MILES*']>50) & (uber_data['MILES*']<100)]
print(filtered_data)
```

```python
filtered_data=uber_data[uber_data['START*']=='Lahore']
print(filtered_data)
```

```python
#ONLY A PARTICULAR CITY SHOULD BE FILTERED AND PRINT ONLY THE START CITY WHICH IS FILTERED
filtered_data=uber_data[uber_data['START*']=='Lahore']
print(filtered_data['START*'])
```

```python
#PRINT ONLY THE FILTERED DATA OF MILES COLUMN
filtered_data=uber_data[(uber_data['MILES*']>50) & (uber_data['MILES*']<100)]
print(filtered_data['MILES*'])
```

```python
#Printing only the 3 columns with a specified condition for MILES*
filtered_data=uber_data[(uber_data['MILES*']>50) & (uber_data['MILES*']<100)]
print(filtered_data[['START*','STOP*','MILES*']])
```
